# Resumen de Reglas de Probabilidad

## Definiciones
Los siguientes términos técnicos se utilizan con frecuencia cuando se habla de probabilidad.

- Evento: algo que se puede observar independientemente de si ocurre o no.
- Probabilidad: la posibilidad de que ocurra un evento, en una escala de 0 (no puede ocurrir) a 1 (siempre ocurre). Se denomina P(evento).
- Universo de probabilidad: el espacio de probabilidad en el que todos los eventos que se están considerando pueden ocurrir o no.
- Eventos mutuamente excluyentes: si ocurre un evento, entonces el otro evento no puede ocurrir (por ejemplo, no se puede tirar un dado que tenga un resultado de 5 y 1).
- Eventos independientes: si ocurre un evento, no afecta la probabilidad de que ocurra el otro evento (por ejemplo, el clima no afecta el resultado de una tirada de dados).
- Eventos dependientes: si ocurre un evento, cambia la probabilidad de que ocurra el otro evento (por ejemplo, el clima afecta los resultados del tráfico).
- Probabilidad conjuntiva (también conocida como probabilidad conjunta): la probabilidad de que ocurran todos los eventos.
- Probabilidad disyuntiva: probabilidad de que ocurra al menos un evento.
- Probabilidad condicional: probabilidad de que ocurra un evento, dado que ocurrió otro evento.

## Regla del complemento: probabilidad de que los eventos no ocurran



- Definición: El complemento de A es la probabilidad de que el evento A no ocurra. Se denota A' o Ac.
- Fórmula: P(A') = 1 - P(A)
- Ejemplo: La probabilidad de que el jugador de baloncesto Stephen Curry acierte un tiro de tres puntos es 0,43. El complemento, la probabilidad de que falle, es 1 - 0,43 = 0,57.

## Probabilidades: Probabilidad de que el evento ocurra versus probabilidad de que no ocurra

- Definición: Las probabilidades de que ocurra el evento A son la probabilidad de que el evento ocurra dividida por la probabilidad de que el evento no ocurra.
- Fórmula: Probabilidades(A) = P(A) / P(A') = P(A) / 1 - P(A)
- Ejemplo: Las probabilidades de que el jugador de baloncesto Stephen Curry lance con éxito un tiro de tres puntos son la probabilidad de que anote dividida por la probabilidad de que falle, 0,43 / 0,57 = 0,75.

## Reglas de multiplicación: Probabilidad de que ocurran ambos eventos
### Eventos mutuamente excluyentes



- Definición: La probabilidad de que ocurran dos eventos mutuamente excluyentes es cero.
- Fórmula: P(A ∩ B) = 0
- Ejemplo: Si la probabilidad de que haya sol al mediodía es 0,3 y la probabilidad de que llueva al mediodía es 0,4, la probabilidad de que haya sol y lluvia es 0, ya que estos eventos son mutuamente excluyentes.

### Eventos independientes



- Definición: La probabilidad de que ocurran dos eventos independientes es el producto de las probabilidades de cada evento.
- Fórmula: P(A ∩ B) = P(A) * P(B)
- Ejemplo: Si la probabilidad de que haya sol al mediodía es 0,3 y la probabilidad de que tu equipo de fútbol favorito gane su partido hoy es 0,6, entonces la probabilidad de que haya sol al mediodía y que tu equipo de fútbol favorito gane su partido hoy es 0,3 * 0,6 = 0,18.

#### La falacia conjuntiva
- Definición: La probabilidad de que ocurran ambos eventos es siempre menor o igual que la probabilidad de que ocurra uno de ellos. Es decir, P(A ∩ B) <= P(A), y P(A ∩ B) <= P(B). La falacia conjuntiva se produce cuando no se piensa detenidamente en las probabilidades y se estima que la probabilidad de que ocurran ambos eventos es mayor que la probabilidad de que ocurra uno de ellos.
- Ejemplo: Un ejemplo famoso conocido como "El problema de Linda" proviene de un experimento de investigación de los años 80. Se describió a una persona ficticia:

    - Linda tiene 31 años, es soltera, franca y muy brillante. Se especializó en filosofía. Como estudiante, estaba profundamente preocupada por los problemas de discriminación y justicia social y también participó en manifestaciones antinucleares.
    - Los participantes tenían que elegir qué afirmación tenía una mayor probabilidad de ser verdadera:
        1. Linda es cajera de banco.
        2.  Linda es cajera de banco y participa activamente en el movimiento feminista.
    - Muchos participantes eligieron la falacia conjuntiva y eligieron la opción 2, aunque debe ser menos probable que la opción 1 utilizando la regla de multiplicación.

## Regla de Bayes: Probabilidad de que ocurra un evento dado que ocurrió otro evento

- Definición: Para eventos dependientes, la probabilidad de que ocurra el evento B, dado que ocurrió el evento A, es igual a la probabilidad de que ocurran ambos eventos dividida por la probabilidad de que ocurra el evento A. 
    - De manera equivalente, es igual a la probabilidad de que ocurra el evento A dado que ocurrió el evento B multiplicada por la probabilidad de que ocurra el evento B dividida por la probabilidad de que ocurra el evento A.
- Fórmula: P(B|A) = P(A ∩ B) / P(A) = P(A|B) * P(B) / P(A)
- Ejemplo: Supongamos que es una mañana nublada y desea saber la probabilidad de que llueva hoy. Si la probabilidad de que llueva ese día dada una mañana nublada es 0,6, y la probabilidad de que llueva cualquier día es 0,1. La probabilidad de que un día cualquiera esté nublado es de 0,3, por lo que la probabilidad de que llueva, dada una mañana nublada, es de 0,6 * 0,1 / 0,3 = 0,2. Es decir, si tienes una mañana nublada, es el doble de probable que llueva que si no tuvieras una mañana nublada, debido a la dependencia de los eventos.
