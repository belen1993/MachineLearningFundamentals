# Agrupación (ensemble) de modelos basados ​​en árboles

Los modelos basados ​​en árboles son predictores poderosos individualmente, pero son un poco limitados.

- En particular, cuando los árboles son profundos, tienden a sobreajustar y si son demasiado poco profundos, tienden a subajustar.

Veremos cómo luchar contra esas debilidades construyendo agrupaciones para obtener modelos más poderosos que se generalizan mejor.

Hay muchas estrategias para construir árboles de decisión.

- La primera estrategia se llama **bolsas (baggins) o bootstrap aggregating**, y puede refinarse aún más en los **bosques aleatorios (random forest)**.
- La segunda familia se llama **boosting** y puede refinarse aún más en los **árboles de decisión gradient boosted**.



Veremos que esas dos estrategias pueden usarse tanto para problemas de clasificación como de regresión.

| Bagging | Boosting |
| --- | --- |
| fit trees independently | fit trees sequentially 
| each deep tree overfits  | each shallow tree underfits 
| averaging the tree predictions reduces overfitting  | sequentially adding trees reduces underfitting |


- Gradient boosting tiende a funcionar un poco mejor que bagging y random forest. 
- Los árboles poco profundos predicen más rápido.
