## ¿Qué es una red neuronal?

Combinemos nuestros cuatro intentos de adivinación en un gran diagrama:



¡Esta es una red neuronal!
- Cada nodo sabe cómo tomar un conjunto de entradas, aplicarles pesos y calcular un valor de salida.
- Al encadenar muchos de estos nodos, podemos modelar funciones complejas.

En otras palabras:
- Hicimos una función de estimación simple que toma un conjunto de entradas y las multiplica por pesos para obtener una salida (**una neurona**).
- Al encadenar muchas neuronas simples, podemos modelar funciones que son demasiado complicadas para ser modeladas por una sola neurona.

## Redes neuronales con memoria

La red neuronal que hemos visto siempre devuelve la misma respuesta para las mismas entradas.
- No tiene memoria.
- En términos de programación, es un *algoritmo sin estado (stateless)*.



Para problemas más complejos, como generar textos, necesitamos agregar estado a nuestro modelo.
- Cada vez que le pedimos a nuestra red neuronal una respuesta, también guardamos un conjunto de nuestros cálculos intermedios y los reutilizamos la próxima vez como parte de nuestra entrada.
- De esa manera, nuestro modelo ajustará sus predicciones en función de la entrada que ha visto recientemente.



Esta es la idea básica de una Red Neuronal Recurrente (Recurrent Neural Network).
- Estamos actualizando la red cada vez que la usamos.
- Esto le permite actualizar sus predicciones en función de lo que vio más recientemente.
- Incluso puede modelar patrones con el tiempo siempre que le demos suficiente memoria.


Pero la parte realmente alucinante es que este algoritmo puede descubrir patrones en cualquier secuencia de datos.
- Puede generar fácilmente recetas de aspecto real o discursos falsos de Obama.
- ¿Pero por qué limitarnos el lenguaje humano? --> Podemos aplicar esta misma idea a cualquier tipo de datos secuenciales que tengan un patrón.
