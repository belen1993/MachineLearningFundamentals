# Overfitting_Underfitting

**¿Qué ajuste prefieres?**



Variando la complejidad del modelo:



**Overfit**: modelo muy complejo



**Undefit**: modelo muy simple



**Conclusiones:**
- Los modelos demasiado complejos para los datos  **sogreajustan**:
    - explican demasiado bien los datos que han visto
    - pero no generalizan
- Los modelos demasiado simples para los datos **subajustan**:
    - No capturan ruido
    - Están limitados por su expresividad
¿Cómo encontrar la compensación correcta?
