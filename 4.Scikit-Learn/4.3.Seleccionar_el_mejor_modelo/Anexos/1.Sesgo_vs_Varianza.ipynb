# Sesgo (Bias) versus Varianza




**Sesgo alto == Subajuste:**
- Errores de predicción sistemáticos
- El modelo prefiere ignorar algunos aspectos de los datos
- Modelos mal especificados

**Alta varianza == Sobreajuste:**
- Errores de predicción sin una estructura obvia
- Pequeño cambio en el conjunto de entrenamiento, gran cambio en el modelo
- Modelos inestables

> El sesgo puede provenir de la elección de la familia modelo.
