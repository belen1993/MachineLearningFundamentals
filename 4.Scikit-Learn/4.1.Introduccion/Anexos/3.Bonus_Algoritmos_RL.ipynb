# Algoritmos Reinforcement Learning

## Aprendizaje por refuerzo
El **aprendizaje por refuerzo (Reinforcement learning)** es el tipo de aprendizaje automático más cercano a cómo aprenden los humanos. El algoritmo o agente utilizado aprende 
interactuando con su entorno y obteniendo una recompensa positiva o negativa. Los algoritmos más comunes incluyen "temporal difference", "deep adversarial networks" y Q-learning.

Es una técnica de aprendizaje automático en la que un agente debe decidir qué acciones debe realizar para realizar una tarea que se le ha asignado de forma más eficaz. 
- Para ello, se asignan recompensas a las diferentes acciones que el agente puede realizar en diferentes situaciones o estados del entorno. 
- Inicialmente, el agente no tiene idea de cuáles son las mejores o correctas acciones. 
- Mediante el aprendizaje por refuerzo, explora sus opciones de acción mediante ensayo y error y determina el mejor conjunto de acciones para completar la tarea asignada.

Requiere varios componentes clave:

- **Agente**: Es el “quién” o el sujeto del proceso, que realiza diferentes acciones para realizar una tarea que le ha sido asignada.
- **Entorno**: es el “lugar” o la situación en la que se encuentra el agente.
- **Acciones**: es el “qué” o los pasos que un agente debe seguir para alcanzar el objetivo.
- **Recompensas**: es la retroalimentación que recibe un agente después de realizar una acción.

### Tipos de aprendizaje por refuerzo
- Aprendizaje por refuerzo basado en modelos
- Aprendizaje por refuerzo sin modelo

Otra clasificación:
- Aprendizaje por refuerzo basado en valores: el agente se centra en aprender una función de valor que estima la recompensa acumulada esperada (rendimiento) de cada estado o par estado-acción. La política se deriva indirectamente seleccionando acciones que maximicen este valor estimado.
- Aprendizaje por refuerzo basado en políticas: se centran en aprender la política directamente sin estimar una función de valor. El agente optimiza la política ajustando sus parámetros para maximizar el rendimiento esperado.

### Algoritmos basados ​​en valores
- Aprendizaje Q: Este algoritmo de RL sin modelo y sin políticas se centra en proporcionar pautas al agente sobre qué acciones debe realizar y bajo qué circunstancias para obtener la recompensa. El algoritmo utiliza tablas Q en las que calcula las recompensas potenciales para diferentes pares de estado-acción en el entorno. La tabla contiene valores Q que se actualizan después de cada acción durante el entrenamiento del agente. Durante la ejecución, el agente vuelve a esta tabla para ver qué acciones tienen el mejor valor.
- Redes Q profundas (DQN): Las redes Q profundas, o deep q-learning, funcionan de manera similar al q-learning. La principal diferencia es que el algoritmo en este caso se basa en redes neuronales.
- SARSA (state-action-reward-state-action): es un algoritmo de RL basado en políticas que utiliza la acción actual de la política actual para aprender el valor.

### Algoritmos basados ​​en políticas
Estos algoritmos actualizan directamente la política para maximizar la recompensa. Existen diferentes algoritmos basados ​​en gradientes de políticas: 
- REINFORCE
- optimización de políticas proximales
- optimización de políticas de regiones de confianza
- algoritmos actor-crítico
- algoritmo actor-crítico de ventaja
- gradiente de políticas determinista profundo (DDPG)
- DDPG con retardo doble.
